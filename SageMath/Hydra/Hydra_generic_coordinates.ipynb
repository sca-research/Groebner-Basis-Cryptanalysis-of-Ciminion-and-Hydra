# Hydra Generic Coordinates
Verfifies that a Hydra instance is in generic coordiantes for an arbitrary number of rounds and arbitrary number of samples.

```python
load("Hydra.sage")
load("Hydra_polynomial_model.sage")
load("../utilities.sage")
```

## Hydra Instance

```python
hydra = Hydra(info_level=1)
```

## m = 2

```python
m = 2

polys = generate_Hydra_polynomials_m_samples(hydra=hydra, m=m)
polys = transform_Hydra_polynomial_system(hydra, polys, m)
```

```python
is_in_generic_coordinates_Hydra_polynomials_m_samples(hydra=hydra, m=m)
```

```python
n_vars = number_of_non_linear_variables_Hydra_polynomial_system(hydra, polys, m, transformed=True)
n_vars
```

```python
len([poly for poly in polys if poly.degree() > 1])
```

```python
macaulay_bound(polys, n_vars=n_vars)
```
